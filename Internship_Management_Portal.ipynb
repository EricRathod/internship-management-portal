```python
import sqlite3

conn = sqlite3.connect("internship_portal.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_name TEXT,
    email TEXT,
    program TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS companies(
    company_id INTEGER PRIMARY KEY AUTOINCREMENT,
    company_name TEXT,
    industry TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS applications(
    application_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_name TEXT,
    company_name TEXT,
    status TEXT
)
""")

conn.commit()
conn.close()

print("Internship Portal Database Created Successfully")
```

Adding Student Records

Search Internship Application

```python
import sqlite3
import pandas as pd

student_name = input("Enter Student Name: ")

conn = sqlite3.connect("internship_portal.db")

query = """
SELECT *
FROM applications
WHERE student_name LIKE ?
"""

result = pd.read_sql_query(
    query,
    conn,
    params=('%' + student_name + '%',)
)

conn.close()

result
```

Update Application Status

```python
import sqlite3

application_id = int(input("Enter Application ID: "))
new_status = input("Enter New Status: ")

conn = sqlite3.connect("internship_portal.db")
cursor = conn.cursor()

cursor.execute("""
UPDATE applications
SET status = ?
WHERE application_id = ?
""", (new_status, application_id))

conn.commit()
conn.close()

print("Application Status Updated Successfully")
```

Application Status Report

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("internship_portal.db")

query = """
SELECT
status,
COUNT(*) AS total_applications
FROM applications
GROUP BY status
"""

report = pd.read_sql_query(query, conn)

conn.close()

report
```
